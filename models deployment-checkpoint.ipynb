```python
import pandas as pd
import numpy as np
import joblib
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.metrics import roc_auc_score, roc_curve

SEED = 42
np.random.seed(SEED)

# 1. Load Data
print("Loading data...")
df = pd.read_csv('Churn_Modelling.csv')
df = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'])

X = df.drop('Exited', axis=1)
y = df['Exited']

# 2. Preprocessing Setup
num_features = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts',
                'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
cat_features = ['Geography', 'Gender']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, cat_features)
    ], remainder='drop')

# 3. Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED, stratify=y)

# 4. Define Models
models = {
    'Logistic Regression': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(random_state=SEED, max_iter=1000, C=1, solver='saga'))
    ]),
    'KNN': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', KNeighborsClassifier(n_neighbors=3, weights='uniform', metric='minkowski'))
    ]),
    'Random Forest': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(
            n_estimators=100, max_depth=10, min_samples_split=10,
            min_samples_leaf=2, class_weight='balanced', random_state=SEED, n_jobs=-1))
    ]),
    'XGBoost': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', xgb.XGBClassifier(
            n_estimators=200, max_depth=6, learning_rate=0.1,
            subsample=0.9, colsample_bytree=0.8, gamma=0.1,
            reg_alpha=1.0, eval_metric='logloss', random_state=SEED, n_jobs=-1))
    ]),
    'SVM': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', SVC(kernel='rbf', C=1, probability=True, random_state=SEED))
    ])
}

# 5. Train and Save
os.makedirs('saved_models', exist_ok=True)
metrics_list = []
roc_data = {}

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    
    # Predictions for metrics
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Calculate ROC Curve data
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc_val = roc_auc_score(y_test, y_prob)
    roc_data[name] = {'fpr': fpr, 'tpr': tpr, 'auc': auc_val}
    
    # Collect Metrics
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    metrics_list.append({
        'Model': name,
        'Accuracy': round(accuracy_score(y_test, y_pred), 4),
        'Precision': round(precision_score(y_test, y_pred), 4),
        'Recall': round(recall_score(y_test, y_pred), 4),
        'F1-Score': round(f1_score(y_test, y_pred), 4),
        'ROC-AUC': round(auc_val, 4)
    })
    
    # Save Model
    filename = name.replace(" ", "_").lower() + "_model.pkl"
    joblib.dump(model, f'saved_models/{filename}')
    print(f"Saved {filename}")

# 6. Save Supporting Files
joblib.dump(pd.DataFrame(metrics_list), 'saved_models/metrics.pkl')
joblib.dump(roc_data, 'saved_models/roc_data.pkl')
joblib.dump(preprocessor, 'saved_models/preprocessor.pkl')
joblib.dump(num_features, 'saved_models/num_features.pkl')
joblib.dump(cat_features, 'saved_models/cat_features.pkl')
joblib.dump(X_test, 'saved_models/X_test.pkl')
joblib.dump(y_test, 'saved_models/y_test.pkl')

print("\nAll models and artifacts saved successfully!")
```
